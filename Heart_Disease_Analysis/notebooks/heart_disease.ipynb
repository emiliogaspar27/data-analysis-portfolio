# Coletando os INSIGHTS

Idade média de casos positivos para doenças cardíacas

```python
Idade_media_feminina = df[(df['sex'] == 0) & (df['target'] == 1)]['age'].mean()
Idade_media_masculina = df[(df['sex'] == 1) & (df['target'] == 1)]['age'].mean()
```

```python
print(f'A idade média para o sexo femimino em que se registrou casos positivos é de {round(Idade_media_feminina)} anos de idade')
print(f'A idade média para o sexo masculino em que se registrou casos positivos é de {round(Idade_media_masculina)} anos de idade')
```

Dor peitoral mais frequente

```python
print(f"A dor mais frequente é a {df['cp'].value_counts().index[0]} com uma percentagem de {round(dor_mais_frequente)}% no conjunto de dados.")
```
