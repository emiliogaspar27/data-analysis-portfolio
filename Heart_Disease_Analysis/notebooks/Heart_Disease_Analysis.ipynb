- #### Quais são os principais fatores de risco combinados para doença cardíaca (idade, sexo, colesterol, pressão, etc.)?

Os principais fatores de risco combinados para doença cardíaca incluem o tipo de dor no peito, frequência cardíaca máxima, inclinação do segmento ST, número de vasos principais coloridos e tipo de thalassemia, que apresentam as associações mais fortes com a presença de doença. Variáveis como idade, sexo, pressão arterial, colesterol, açúcar no sangue e resultados do ECG têm efeitos isolados mais fracos, mas podem contribuir quando analisadas em conjunto com os fatores principais. Em resumo, a combinação dessas variáveis fornece a melhor predição do risco de doença cardíaca neste dataset.
